```python
import tensorflow as tf

class myCallback(tf.keras.callbacks.Callback):
  def on_epoch_end(self, epoch, logs={}):
    if(logs.get('accuracy')>0.95):
      print("\nReached 95% accuracy so cancelling training!")
      self.model.stop_training = True

callbacks = myCallback()

data = tf.keras.datasets.fashion_mnist

(training_images, training_labels), (test_images, test_labels) = data.load_data()

# normalize all images to represent each pixel with 1 or 0 - improve perf
training_images = training_images / 255.0
test_images = test_images / 255.0

model = tf.keras.models.Sequential([
  tf.keras.layers.Flatten(input_shape=(28,28)),
  # rectified linear unit returns values > 0
  tf.keras.layers.Dense(128, activation=tf.nn.relu),
  tf.keras.layers.Dense(10, activation=tf.nn.softmax)
])

# adam slight perf upgrade from sgd
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
#callback on_epoch_end will fire at 95% accuracy ending epochs early
model.fit(training_images, training_labels, epochs=50, callbacks=[callbacks])

print('evaluating test images and labels:')

model.evaluate(test_images, test_labels)

classifications = model.predict(test_images)
print(classifications[1])
print(test_labels[1])
```
